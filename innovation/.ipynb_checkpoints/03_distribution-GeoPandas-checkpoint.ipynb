## 2.1 매출액 만원단위 변경

```python
df_dist['AMT_VAL'] = df_dist['AMT_VAL'].apply(lambda x : x / 10000)
df_dist['PROD_TP_AMT_VAL'] = df_dist['PROD_TP_AMT_VAL'].apply(lambda x : np.round(x/10000, 3))
```

```python
df_dist.head()
```

## 2.2 구코드 및 동코드 지도 데이터 코드와 맞추기

```python
df_dist['BOR_CD'] = df_dist['BOR_CD'].apply(lambda x : str(x) + "0")
df_dist.head()
```

```python
df_dist['ADMD_CD'] = df_dist['ADMD_CD'].apply(lambda x : str(x) + "00")
df_dist.head()
```

## 2.3 매출 데이터 구, 동 코드 데이터프레임 생성
