# 미세먼지 데이터 로드

```python
dust_jr = pd.read_csv('./weather_jr.csv')
```

```python
dust_nw = pd.read_csv('./weather_nw.csv')
```

```python
dust_mean = pd.concat([dust_jr,dust_nw], axis=1, join='inner')
dust_mean.head()
```

```python
dust_mean.columns
```

```python
dust_mean = dust_mean.drop(['serial','flag','co2','noise','temp','humi','V11','vocs'], axis=1)
```

# 미세먼지,초미세먼지 이상치 제거

```python
import numpy as np

#미세먼지 이상치값 없애는 함수 만들기
def reject_outliers(dust_mean):
    q1, q3 = np.percentile(dust_mean['pm10'], [25, 75])
    iqr = q3 - q1
    lower = q1 - (iqr * 3)
    upper = q3 + (iqr * 3)
    dust_mean = dust_mean[(dust_mean["pm10"]>lower) & (dust_mean["pm10"]<upper)]
    return dust_mean

#초미세먼지 이상치값 없애는 함수 만들기
def reject_outliers2(dust_mean):
    q1, q3 = np.percentile(dust_mean['pm25'], [25, 75])
    iqr = q3 - q1
    lower = q1 - (iqr * 3)
    upper = q3 + (iqr * 3)
    dust_mean = dust_mean[(dust_mean["pm25"]>lower) & (dust_mean["pm25"]<upper)]
    return dust_mean
```
