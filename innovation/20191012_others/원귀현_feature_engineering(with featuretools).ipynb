```python
num=0
for i in jong:
    print("------"+str(num)+"-------")
    if num==0:
        data1=pd.read_csv("./data/종로구/"+i+".csv")
        date1 = '2018-04-01 00:00'
        date2 = '2019-03-31 23:59'
        mydates = pd.date_range(date1, date2,freq='1min').tolist()
        total_data=pd.DataFrame({"tm":mydates})
        data1.tm=pd.to_datetime(data1.tm.apply(str))
        total_data=total_data.merge(data1,how="left")
        a1=total_data.tm.apply(str)
        total_data['tm2']=pd.to_datetime(a1.str.slice(0,10), format='%Y-%m-%d')
        num=num+1
            
    else:
        data2=pd.read_csv("./data/종로구/"+i+".csv")
        date1 = '2018-04-01 00:00'
        date2 = '2019-03-31 23:59'
        mydates = pd.date_range(date1, date2,freq='1min').tolist()
        total_data2=pd.DataFrame({"tm":mydates})
        data2.tm=pd.to_datetime(data2.tm.apply(str))
        total_data2=total_data2.merge(data2,how="left")
        a1=total_data2.tm.apply(str)
        total_data2['tm2']=pd.to_datetime(a1.str.slice(0,10), format='%Y-%m-%d')
        total_data=pd.concat([total_data,total_data2],axis=0)
        num=num+1
```

```python
####연구(완료)
total_data=total_data.drop(['Unnamed: 10'],axis=1)
key1=list(set(total_data.tm2))
df1=pd.DataFrame({"tm2":key1})
df1
es=ft.EntitySet(id='sub1')
es.entity_from_dataframe(entity_id='data',dataframe=total_data,index='ind1')
es.entity_from_dataframe(entity_id='key',dataframe=df1,index='tm2')
r1=ft.Relationship(es['key']['tm2'],es["data"]['tm2'])
es=es.add_relationship(r1)
```

```python
mat1,feature1=ft.dfs(entityset=es,target_entity='key',agg_primitives=["min", "max","mean",'std','median'])
```
