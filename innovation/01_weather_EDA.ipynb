## 1.2 컬럼별 값의 빈도수

```python
w_j_cols = list(df_w_j.columns)

# value count 조회 비대상 컬럼 제외
w_j_cols.remove("tm")
w_j_cols.remove("serial")
w_j_cols.remove("date")
w_j_cols.remove("V11")

w_j_cols
```

```python
# 각 컬럼의 값별 빈도수 데이터프레임을 생성하는 함수 정의
def make_vc_df(df, col) :
    
    df_vc = df[col].value_counts().reset_index()
    df_vc.rename(columns={"index": "value", col: "count"}, inplace=True)
    df_vc.sort_values(by="value", inplace=True)
    
    return df_vc
```

### 1.2.1 flag

```python
df_flag_j = make_vc_df(df_w_j, 'flag')
df_flag_j
```

```python
df_flag_n = make_vc_df(df_w_n, 'flag')
df_flag_n
```

```python
plt.figure(figsize=(18, 6))

plt.subplot(121)
g_flag_j = sns.barplot(x="value", y="count", data=df_flag_j)
g_flag_j.set_title("Flag's Frequency of value in Jongno", fontsize=16)

plt.subplot(122)
g_flag_n = sns.barplot(x='value', y='count', data=df_flag_n)
g_flag_n.set_title("Flag's Frequency of value in Nowon", fontsize=16)

plt.show()
```
