## 1.3 지역별 월별 매출

```python
df_dist_j2 = df_dist_j[['OPER_DT','PROD_TP_AMT_VAL']]
df_dist_j2.head()
```

```python
df_dist_j2['OPER_DT'].astype(str).iloc[0,6]
```

```python
df_dist_j2['OPER_YM'] = df_dist_j2['OPER_DT'][:6]
df_dist_j2.head()
```

```python
df_dist_j3 = df_dist_j2.groupby(df_dist_j2.OPER_DT).sum().reset_index()
df_dist_j3.head()
```

```python
sns.barplot(x='OPER_DT', y='PROD_TP_AMT_VAL', data=df_dist_j3)
plt.show()
```

```python
sns.distplot(df_dist_j3['PROD_TP_AMT_VAL'])
plt.show()
```

```python
df_dist_n2 = df_dist_j[['OPER_DT','PROD_TP_AMT_VAL']]

df_dist_n3 = df_dist_n2.groupby(df_dist_n2.OPER_DT).sum().reset_index()
df_dist_n3.head()
```

```python
sns.barplot(x='OPER_DT', y='PROD_TP_AMT_VAL', data=df_dist_n3)
plt.show()
```

```python
sns.distplot(df_dist_n3['PROD_TP_AMT_VAL'])
plt.show()
```
